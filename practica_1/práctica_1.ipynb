1. Crear el KPI de "Valor de Cliente" (Customer Score)

No todos los clientes valen lo mismo. Crea una nueva columna llamada puntuacion_lealtad siguiendo esta lógica simple:

    Puntos: visitas_año multiplicado por 0.7 + suscrito multiplicado por 10.

    Meta: Identifica el id_cliente con la puntuación más alta.

2. Segmentación de Clientes (Etiquetado)

Usa la columna gasto_anual para clasificar a los 100 clientes en tres categorías:

    "Bajo Gasto": Menos de $10,000.

    "Gasto Medio": Entre $10,000 y $25,000.

    "VIP": Más de $25,000.

    Meta: ¿Cuántos clientes "VIP" tenemos en total?

3. Análisis de Abandono (Churn Risk)

Crea una columna booleana (True/False) llamada riesgo_abandono.

    Un cliente tiene riesgo si han pasado más de 150 días desde su última visita (dias_ultima_visita).

    Meta: ¿Qué estado (CDMX, Jalisco, etc.) tiene el mayor número de clientes en riesgo?

¿Cuántos clientes VIP encontraste?

¿Cuál es el estado con más clientes en riesgo de abandono?

```python
df['puntuacion_lealtad'] = (df['visitas_año'] * 0.7) + (df['suscrito'] * 10)
#df
dfPuntuacion = df[['id_cliente', 'puntuacion_lealtad']]
dfPuntuacion.sort_values(by='puntuacion_lealtad', ascending=False)
```

```python

def segmentarClientes(gasto_anual):
    if gasto_anual <= 10000:
        res = 'Bajo Gasto'
        return res
    elif gasto_anual >= 10000 and gasto_anual <= 25000:
        res = 'Gasto Medio'
        return res
    elif gasto_anual >= 25000:
        res = 'VIP'
        return res
    

df['etiqueta'] = df['gasto_anual'].apply(segmentarClientes)


#clientes_vip = df[(df['gasto_anual'] >= 25000)]

clientes_vip = df[(df['etiqueta'] == 'VIP')]
print("Se tienen",len(clientes_vip), "clientes VIP")
```

```python
import plotly.express as px

df_agrupado = df.groupby('estado').agg({
    'gasto_anual' : 'sum'
}).reset_index()


fig = px.bar(
   df_agrupado,
   x = 'estado',
   y = 'gasto_anual',
   title= 'Total de gasto anual por estado',
   labels = {
       'estado' : 'Estado',
       'gasto_anual' : 'Gasto anual'
       
   },
   template="plotly_dark",
)

fig.show()

#df_agrupado
```

```python
df_edades = df[['edad', 'gasto_anual', 'estado']]
df_edades

fig2 = px.scatter(
    df_edades,
    x='edad',
    y='gasto_anual',
    color='estado',
    title='Relacion del gasto anual respecto a la edad del cliente', 
    labels={
        'edad' : 'Edad',
        'gasto_anual' : 'Gasto anual'
    },
    template="plotly_dark"
)

fig2.show()

#df_edades['edad'].corr(df_edades['gasto_anual'])
```
