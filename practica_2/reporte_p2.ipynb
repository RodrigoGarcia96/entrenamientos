## Sección 4

- Propón una forma de segmentar a los pacientes según su costo de tratamiento.
- ¿Qué criterio usarías para definir pacientes de alto costo?


Para resolver la primera parte, lo que decido es calcular los cuartiles para segregarlos de alto costo, medio costo y bajo costo de la variable "costo_tratamiento". Los cuartiles me ayudan a ver los valores de 25%, 50% y 75% para su clasificación usando la variable de "costo_tratamiento".

```python
df['costo_tratamiento'].describe()
```

```python
q1 = df['costo_tratamiento'].quantile(0.25)
q3 = df['costo_tratamiento'].quantile(0.75)

def segmentar_costo(costo_tratamiento):
    if costo_tratamiento <= q1:
        res = 'bajo'
        return res
    elif costo_tratamiento <= q3:
        res = 'medio'
        return res
    else:
        res = 'alto'
        return res


df['segmento_costo'] = df['costo_tratamiento'].apply(segmentar_costo)
df.head()
#print(q1)
```

## Sección 6 

- ¿Qué tipo de gráfica usarías para analizar la relación entre días hospitalizado y costo del tratamiento?
- ¿Qué insight de negocio o clínico podrías obtener de esa visualización?

Yo use un gráfico de burbujas para ver la relación entre los días que estaban en el hospital y el costo del tratamiento, también quería ver si los pacientes más grandes eran los que resibían los tratamientos más costosos, a lo que hago la siguiente visualización.

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "notebook_connected"

fig1 = px.scatter(
    df,
    x= 'dias_hospitalizado',
    y='costo_tratamiento',
    color='segmento_costo',
    color_discrete_map={
                 'alto': "#EF553B",
                 'medio' : '#FECB52',
                 'bajo': "#00CC96"
                 
             },
    template='plotly_dark',
    title='Relación días hospitalizado y costo del tratamiento',
    labels={
        'costo_tratamiento' : 'Costo del tratamiento',
        'segmento_costo' : 'Costo por segmento',
        'dias_hospitalizado' : 'Dias hospitalizado',
        'alto' : 'alto costo'
    },
    category_orders={"segmento_costo": ["alto", "medio", "bajo"]},
    size = 'edad'
    )

fig1.update_xaxes(
    range=[0, 20],      
    dtick=1             
)

fig1.show()
```

Aqui podemos ver que a mayor días que pasan, incrementa el costo del tratamiento.
